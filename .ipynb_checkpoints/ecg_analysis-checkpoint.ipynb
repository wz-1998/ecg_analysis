## 1. Data Acquisition & Preparation

**Source:** MIT-BIH Arrhythmia Database (PhysioNet)

- Downloaded 44 patient records (excluded paced rhythms) via WFDB Python package

- Patient-wise split (70% train / 10% validation / 20% test) to prevent data leakage

- Records included: ['100', '101', ..., '234'] (Full list in code)

**Ethical Considerations:**

- PhysioNet credentialed access

- HIPAA-compliant de-identified data

```python
import wfdb
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.utils import shuffle

# Configuration
RECORD_NUMBERS = [
    '100', '101', '103', '105', '106', '108', '109', '111', '112', '113', '114',
    '115', '116', '117', '118', '119', '121', '122', '123', '124', '200', '201',
    '202', '203', '205', '207', '208', '209', '210', '212', '213', '214', '215',
    '219', '220', '221', '222', '223', '228', '230', '231', '232', '233', '234'
]  # 44 records excluding paced beats

CLASS_MAPPING = {
    'N': 0,  # Normal
    'V': 1,  # Premature ventricular contraction
}

SEGMENT_LENGTH = 360  # 1 second at 360 Hz
TEST_SIZE = 0.2
VAL_SIZE = 0.1
RANDOM_STATE = 12

def load_data(record_numbers):
    """Load ECG segments and labels from specified records"""
    segments = []
    labels = []
    
    for record in record_numbers:
        try:
            # Download data from PhysioNet
            signals, _ = wfdb.rdsamp(record, pn_dir='mitdb', channels=[0])
            ann = wfdb.rdann(record, 'atr', pn_dir='mitdb')
            
            ecg = signals[:, 0]  # First lead
            
            # Extract segments around each R-peak
            for idx, symbol in zip(ann.sample, ann.symbol):
                if symbol in CLASS_MAPPING:
                    start = idx - SEGMENT_LENGTH//2
                    end = idx + SEGMENT_LENGTH//2
                    
                    # Handle boundary conditions with padding
                    if start < 0 or end > len(ecg):
                        pad_left = max(0, -start)
                        pad_right = max(0, end - len(ecg))
                        segment = np.pad(ecg[max(0, start):min(end, len(ecg))],
                                       (pad_left, pad_right), mode='constant')
                    else:
                        segment = ecg[start:end]
                        
                    segments.append(segment)
                    labels.append(CLASS_MAPPING[symbol])
                    
        except Exception as e:
            print(f"Error processing record {record}: {e}")
    
    return np.array(segments), np.array(labels)

def normalize(segments):
    """Z-score normalization for each segment"""
    return np.array([(s - np.mean(s)) / np.std(s) for s in segments])
```

```python
# Split records into train, validation, and test sets
train_records, test_records = train_test_split(
    RECORD_NUMBERS, test_size=TEST_SIZE, random_state=RANDOM_STATE
)
train_records, val_records = train_test_split(
    train_records, test_size=VAL_SIZE/(1-TEST_SIZE), random_state=RANDOM_STATE
)
```

## 2. Preprocessing Pipeline

**Key Steps:**

1. **R-Peak Centering:** Extracted 1-second segments (360 samples) around annotated beats
2. **Class Filtering:** Kept only Normal (N) and PVC (V) beats per AAMI standards
3. **Z-Score Normalization:** (x - μ)/σ per segment to handle amplitude variation
4. **Padding:** Zero-padded edge segments to maintain 360-sample length

**Class Distribution:**
| Class | Train	| Validation | Test |
| --- | --- | --- | --- |
| N	| 55,635 | 5,740 | 13,171 |
| V	| 4,094	| 470 | 2,339 |

```python
# Load and process datasets
print("Loading training data...")
X_train, y_train = load_data(train_records)
X_train = normalize(X_train)
X_train, y_train = shuffle(X_train, y_train, random_state=RANDOM_STATE)

print("Loading validation data...")
X_val, y_val = load_data(val_records)
X_val = normalize(X_val)
X_val, y_val = shuffle(X_val, y_val, random_state=RANDOM_STATE)

print("Loading test data...")
X_test, y_test = load_data(test_records)
X_test = normalize(X_test)
X_test, y_test = shuffle(X_test, y_test, random_state=RANDOM_STATE)

# Print dataset information
print("\nDataset shapes:")
print(f"Train: {X_train.shape}, {y_train.shape}")
print(f"Validation: {X_val.shape}, {y_val.shape}")
print(f"Test: {X_test.shape}, {y_test.shape}")

print("\nClass distribution (train):")
print(np.unique(y_train, return_counts=True))
print("\nClass distribution (val):")
print(np.unique(y_val, return_counts=True))
print("\nClass distribution (test):")
print(np.unique(y_test, return_counts=True))
```

## Model Architecture

**1D CNN for Temporal Pattern Recognition**

|  Layer (type) | Output Shape | Param #  |
|---|---|---|
| conv1d (Conv1D)  | (None, 360, 32)  | 192  |                                                           
| batch_normalization (BatchNormalization) | (None, 360, 32) | 128 |
| max_pooling1d (MaxPooling1D) | (None, 180, 32) |  0  |
| dropout (Dropout)    |       (None, 180, 32)     |      0         |                                               
| conv1d_1 (Conv1D)    |       (None, 180, 64)       |    6208      |                                                            
| batch_normalization_1 (BatchNormalization) | (None, 180, 64)     |     256      | 
| max_pooling1d_1 (MaxPooling 1D) | (None, 90, 64)     |      0     |    
| dropout (Dropout)    |       (None, 90, 64)     |      0         | 
| flatten (Flatten)     |      (None, 5760)      |        0         |
| dense (Dense)       |        (None, 64)        |        368,704 |
| dropout_1 (Dropout)    |     (None, 64)       |         0    |
| dense_1 (Dense)        |     (None, 2)     |            130 |

**Design Rationale:**
- Small kernel sizes (5-3) to capture local ECG morphologies
- Progressive downsampling preserves temporal relationships
- Dropout layers (20-40%) prevent overfitting on imbalanced data

```python
from tensorflow.keras import layers, models
from sklearn.metrics import classification_report, confusion_matrix

# Ensure input data has channel dimension
if len(X_train.shape) < 3:
    X_train = np.expand_dims(X_train, axis=-1)
    X_val = np.expand_dims(X_val, axis=-1)
    X_test = np.expand_dims(X_test, axis=-1)

def create_baseline_cnn(input_shape, num_classes):
    model = models.Sequential([
        layers.Input(shape=input_shape),
        
        # First convolutional block
        layers.Conv1D(filters=32, kernel_size=5, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(pool_size=2),
        layers.Dropout(0.2),
        
        # Second convolutional block
        layers.Conv1D(filters=64, kernel_size=3, activation='relu', padding='same'),
        layers.BatchNormalization(),
        layers.MaxPooling1D(pool_size=2),
        layers.Dropout(0.3),
        
        # Classification head
        layers.Flatten(),
        layers.Dense(64, activation='relu'),
        layers.Dropout(0.4),
        layers.Dense(num_classes, activation='softmax')
    ])
    return model
```

## 4. Data Augmentation Strategy

Use Generator class to apply random transformations of training data to simulate noise and prevent overfitting.

**ECGDataGenerator Class Features:**

- Time Warping: ±20% stretch/compression using linear interpolation
- Additive Noise: Gaussian (μ=0, σ=0.01×segment σ)
- Amplitude Scaling: 0.8×-1.2× random gain

```python
from sklearn.utils.class_weight import compute_class_weight

# Custom data augmentation generator
class ECGDataGenerator(tf.keras.utils.Sequence):
    def __init__(self, X, y, batch_size=64, augment=False):
        self.X = X
        self.y = y
        self.batch_size = batch_size
        self.augment = augment
        self.class_indices = {0: np.where(y == 0)[0],
                              1: np.where(y == 1)[0]}
        
    def __len__(self):
        return int(np.ceil(len(self.X) / self.batch_size))
    
    def __getitem__(self, idx):
        batch_X = self.X[idx*self.batch_size : (idx+1)*self.batch_size]
        batch_y = self.y[idx*self.batch_size : (idx+1)*self.batch_size]
        
        if self.augment:
            batch_X = np.array([self.augment_ecg(x, y) for x, y in zip(batch_X, batch_y)])
            
        return batch_X, batch_y
    
    def augment_ecg(self, segment, label):
        if label == 0:  # Only augment minority classes
            return segment
        
        # Time warping
        if np.random.rand() < 0.7:
            warp_factor = np.random.uniform(0.8, 1.2)
            new_length = int(len(segment) * warp_factor)
            x_original = np.linspace(0, 1, len(segment))
            x_new = np.linspace(0, 1, new_length)
            segment = np.interp(x_new, x_original, segment.flatten()).reshape(-1, 1)
            if len(segment) < 360:
                segment = np.pad(segment, ((0, 360-len(segment)), (0, 0)), mode='edge')
            else:
                segment = segment[:360]
        
        # Additive noise
        if np.random.rand() < 0.5:
            noise = np.random.normal(0, 0.02*np.std(segment), segment.shape)
            segment += noise
            
        # Amplitude scaling
        if np.random.rand() < 0.3:
            scale = np.random.uniform(0.8, 1.2)
            segment *= scale
            
        return segment


# Calculate class weights
class_weights = compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
class_weight_dict = {i: weight for i, weight in enumerate(class_weights)}
print("Class weights:", class_weight_dict)
```

## 5. Training Protocol

**Imbalanced Learning Approach:**
- Class Weights: Inversely proportional to class frequency
- Early Stopping: Patience=10 epochs (val_loss)

```python
# Model parameters
INPUT_SHAPE = (X_train.shape[1], 1)
NUM_CLASSES = len(np.unique(y_train))

# Create and compile model
model = create_baseline_cnn(INPUT_SHAPE, NUM_CLASSES)
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy',]
)

# Create data generators
train_generator = ECGDataGenerator(X_train, y_train, batch_size=64, augment=True)
val_generator = ECGDataGenerator(X_val, y_val, batch_size=64)

# Callbacks
callbacks = [
    tf.keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, mode='max', restore_best_weights=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, verbose=1)
]

# Train the model
history = model.fit(
    train_generator,
    validation_data=val_generator,
    epochs=100,
    class_weight=class_weight_dict,
    callbacks=callbacks,
    verbose=1
)

# Evaluation
print("\nEvaluating on test set:")
y_pred = model.predict(X_test)
y_pred_classes = np.argmax(y_pred, axis=1)

print("\nClassification Report:")
print(classification_report(y_test, y_pred_classes, target_names=['N', 'V']))

print("\nConfusion Matrix:")
print(confusion_matrix(y_test, y_pred_classes))
```

## 7. Model Interpretation

Grad-CAM Visualization:

```python
from tensorflow.keras.models import Model

def grad_cam(model, sample, class_idx, layer_name='conv1d_1'):
    # create Model that outputs
    # 1) activations from target layer
    # 2) final classification probabilities
    grad_model = Model(
        inputs=model.inputs,
        outputs=[model.get_layer(layer_name).output, model.output]
    )

    # 1) pass ECG sample through model
    # 2) extract conv_outputs (feature maps from target layer) and predictions (class probs)
    # 3) take only the prediction confidence from predicted class (class_channel)
    with tf.GradientTape() as tape:
        last_conv_layer_output, preds = grad_model(np.array([sample]))
        pred_index = tf.argmax(preds[0])
        class_channel = preds[:, pred_index]

    # calculate gradients with respect to each value in conv_outputs
    # average gradients across time steps
    grads = tape.gradient(class_channel, last_conv_layer_output)
    pooled_grads = tf.reduce_mean(grads, axis=(0))
    
    last_conv_layer_output = last_conv_layer_output[0]

    # multiply each feature map (64 total) by corresponding gradient importance weight
    # emphasizes filters that are weighted higher by gradient
    heatmap = last_conv_layer_output * pooled_grads
    heatmap = tf.reduce_mean(heatmap, axis=(1))
    heatmap /= np.max(heatmap)
    heatmap = np.expand_dims(heatmap,0)
    return heatmap, pred_index, class_channel

def plot_gradcam(sample, heatmap):
    plt.figure(figsize=(30,4))
    plt.imshow(np.expand_dims(heatmap,axis=2),cmap='Reds', aspect="auto", interpolation='nearest',extent=[0,360,sample.min(),sample.max()], alpha=0.5)
    plt.plot(sample,'k')
    plt.colorbar()
    plt.title('Grad-CAM Activation vs ECG')
    plt.xlabel('Time Steps')
    plt.ylabel('Amplitude')
    plt.tight_layout()
    plt.show()

# Plot Normal QRS examples
for i in range(0,10):
    class_y = np.where(y_test == 0)[0]
    randi = np.random.randint(0, class_y.shape[0])
    sample_idx = class_y[randi]  # Get random PVC sample
    sample = X_test[sample_idx].squeeze()
    
    # Generate heatmap
    heatmap, pred_idx, class_loss = grad_cam(model, sample, class_idx=1)
    
    # Plot visualization
    print(f"Sample {i}: Model prediction = {pred_idx}, Confidence = {class_loss}, True label = {y_test[sample_idx]}")
    plot_gradcam(sample, heatmap)
```

```python
# Plot PVC (V) examples
for i in range(0,10):
    class_y = np.where(y_test == 1)[0]
    randi = np.random.randint(0, class_y.shape[0])
    sample_idx = class_y[randi]  # Get random PVC sample
    sample = X_test[sample_idx].squeeze()
    
    # Generate heatmap
    heatmap, pred_idx, class_loss = grad_cam(model, sample, class_idx=1)
    
    # Plot visualization
    print(f"Sample {i}: Model prediction = {pred_idx}, Confidence = {class_loss}, True label = {y_test[sample_idx]}")
    plot_gradcam(sample, heatmap)
```

- Minimal attention to P-waves/T-waves in PVC classification
- Aligns with clinical PVC characteristics
